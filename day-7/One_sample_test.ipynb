```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm
```

```python
 df_2015 =pd.read_csv('../dataset/marathon_results_2015.csv')
 df_2016 =pd.read_csv('../dataset/marathon_results_2016.csv')
 df_2017 =pd.read_csv('../dataset/marathon_results_2017.csv')
```

```python
#concat the dataset into single datafraome
df = pd.concat([df_2015,df_2016,df_2017])
df.shape
df['Age'].mean()
```

```python
np.random.seed(4)
age_data = df['Age'].sample(n=30,replace=False)
```

```python
pop_avg_age = 48
alpha = 0.05
t_test,p_value = stats.ttest_1samp(age_data,pop_avg_age)
print(f"{t_test:.4f}")
print(f"{p_value:.4f}")
```

```python
#degree of freedom
degree_of_freedom = len(age_data) - 1
critical = stats.t.ppf(1 - alpha/2,degree_of_freedom)
print(critical)
```

```python
#interpretation

if p_value < alpha:
    print(f"The null hypothesis (mean age = {pop_avg_age}) is rejected")
else:
    print(f"The null hypothesis (mean age = {pop_avg_age}) is can not be rejected")
```

```python
#interpretation criticala value
if t_test > critical:
     print(f"The null hypothesis (mean age = {pop_avg_age}) is rejected")
else:
    print(f"The null hypothesis (mean age = {pop_avg_age}) is can not be rejected")

    
```

```python
plt.figure(figsize=(10,6))
x = np.linspace(-4,4,1000)
y=norm.pdf(x)
plt.plot(x,y,label="Standard Normal Ditribution",color='black')
plt.fill_between(x,0,y,where=((x > -critical) | (x < critical)) ,color='green',alpha=0.7,label="Non Rejection Region")
plt.fill_between(x,0,y,where=((x <= -critical) | (x >= critical)) ,color='red',alpha=0.8,label="Rejection Region")

plt.axvline(critical,linestyle='--',color='black',label=f'Critcal_two_tailed: {critical:.2f}')
plt.axvline(-critical,linestyle='--',color='black',label=f'Critcal_two_tailed: {critical:.2f}')


plt.axvline(t_test,linestyle='--',color='blue',label=f'Z-Statistics: {t_test:.2f}')
plt.title("Right One Tailed z-Test visulization")
plt.xlabel("Z-Score")
plt.ylabel("Probability Distribution")
plt.legend()
plt.show()
plt.tight_layout()
```
