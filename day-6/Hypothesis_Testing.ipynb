```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm
```

Hypothesis
H0: population average tip value is $2.75


```python
mue_0 = 2.75
df = pd.read_csv('../dataset/tips.csv')
tips = df['tip']


```

```python
#1) calcualte the mean and standard deviation of sample data
sample_mean = tips.mean()
sample_std = tips.std()
n =len(tips)
```

```python
# calculate the z-statastics 
standard_err = sample_std/(np.sqrt(n))
z_stat = (sample_mean - mue_0) / standard_err 
```

```python
#calculate critical value for z statistics for alpha = 0.05

alpha = 0.05
critical_two_tailed = norm.ppf(1 - alpha/2)
critical_one_tailed = norm.ppf(1 - alpha)

#p_value 
p_value_two = 2 * (1 - norm.cdf(abs(z_stat)))
p_value_one= 1 - norm.cdf(abs(z_stat))
```

```python
#pring the details 
print(f"Sample Mean : {sample_mean:.2f}")
print(f"Sample std  : {sample_std:.2f}")
print(f"Z-statistic : {z_stat:.2f}")

print(f"Critical two tailed test +/- {critical_two_tailed:.2f}")

print(f"p_value two tailed test {p_value_two:.4f}")
```

```python
#interpret the result for two tailed test
if abs(z_stat) > critical_two_tailed:
     print("Reject the null hypothesis for two tailed test")
else:
     print("failed to Reject the null hypothesis for two tailed test")
```

```python
if p_value_two < alpha:
     print("Reject the null hypothesis for two tailed test")
else:
     print("failed to Reject the null hypothesis for two tailed test")
    
```

```python
plt.figure(figsize=(10,6))
x = np.linspace(-4,4,1000)
y=norm.pdf(x)
plt.plot(x,y,label="Standard Normal Ditribution",color='black')
plt.fill_between(x,0,y,where=((x > -critical_two_tailed) |(x < critical_two_tailed)),color='green',alpha=0.7,label="Non Rejection Region")
plt.fill_between(x,0,y,where=((x <= -critical_two_tailed) |(x >= critical_two_tailed)),color='red',alpha=0.8,label="Rejection Region")

plt.axvline(critical_two_tailed,linestyle='--',color='black',label=f'Critcal_two_tailed: +{critical_two_tailed:.2f}')
plt.axvline(-critical_two_tailed,linestyle='--',color='black',label=f'Critcal_tww_tailed: -{critical_two_tailed:.2f}')

plt.axvline(z_stat,linestyle='--',color='blue',label=f'Z-Statistics: {z_stat:.2f}')
plt.title("Two Tailed z-Test visulization")
plt.xlabel("Z-Score")
plt.ylabel("Probability Distribution")
plt.legend()
plt.show()
plt.tight_layout()
```

```python
print(f"Z-statistic : {z_stat:.2f}")
print(f"Critical one tailed test +/- {critical_one_tailed:.2f}")

print(f"p_value one tailed test {p_value_one:.4f}")
```

```python
#interpretation for left one tailed test

if z_stat < -critical_one_tailed:
     print("Reject the null hypothesis for two tailed test")
else:
     print("Fail to Reject the null hypothesis for two tailed test")
```
